### Задание: Разработка проекта TextEasyProcessorPro на Django

Необходимо разработать проект на Django с названием **TextEasyProcessorPro**, который будет включать в себя следующие элементы:

1. **Базовая страница (главная страница)**:
   - **Левая панель**: Должна содержать боковое меню (drawer menu), аналогичное тому, что используется на сайте [https://chat.deepseek.com/](https://chat.deepseek.com/).
   - **Верхняя панель**: Должна содержать основное меню для навигации по проекту.
   - **Левый блок**: Отдельный блок для отображения важной информации (например, рекламы или уведомлений).
   - **Центральный блок**: Основной блок для отображения контента.
   - **Футер**: Нижняя часть страницы с дополнительной информацией (например, контакты, ссылки на документацию и т.д.).

2. **HTML-формы**:
   Все формы в проекте должны наследоваться от базовой формы, которая определяет общую структуру страницы (включая боковое меню, верхнее меню, левый блок и футер).



### Форма 1: TextClassificationForm

**Описание**:  
Форма предназначена для классификации текстов с использованием различных моделей машинного обучения и методов векторизации. Форма должна содержать следующие поля:

#### Поля формы:

1. **dataset_path_or_name**:
   - **Тип**: `CharField`
   - **Описание**: Поле для указания пути к датасету или выбора названия датасета из предопределенного списка.
   - **Виджет**: `TextInput` (для ввода пути) или `Select` (для выбора из списка).
   - **Начальное значение**: Пустое поле или значение по умолчанию, например, `"default_dataset"`.

2. **dataset_categories**:
   - **Тип**: `MultipleChoiceField`
   - **Виджет**: `CheckboxSelectMultiple`
   - **Описание**: Поле для выбора категорий данных из предопределенного списка (`DATASET_CATEGORIES`).
   - **Начальное значение**: `['sci.med', 'sci.space']`.

3. **sample_size**:
   - **Тип**: `IntegerField`
   - **Описание**: Поле для указания размера выборки данных, используемой для обучения модели.
   - **Минимальное значение**: `100`.
   - **Максимальное значение**: `100 000`.
   - **Начальное значение**: `500`.

4. **models**:
   - **Тип**: `MultipleChoiceField`
   - **Виджет**: `CheckboxSelectMultiple`
   - **Описание**: Поле для выбора моделей машинного обучения из предопределенного списка (`MODEL_CHOICES`).
   - **Начальное значение**: `['logistic_regression', 'random_forest']`.

5. **vectorizer**:
   - **Тип**: `ChoiceField`
   - **Описание**: Поле для выбора метода векторизации текста из предопределенного списка (`VECTORIZER_CHOICES`).
   - **Начальное значение**: `'tf_idf'`.

6. **hyperparameter_tuning**:
   - **Тип**: `BooleanField`
   - **Описание**: Флаг, указывающий, нужно ли выполнять подбор гиперпараметров для выбранных моделей.
   - **Начальное значение**: `True`.

7. **output_directory**:
   - **Тип**: `CharField`
   - **Описание**: Поле для указания директории, в которую будут сохраняться обученные модели.
   - **Начальное значение**: `"classificationmodels"`.

8. **train_button** и **test_button**:
   - **Описание**:  
     - **train_button**: Кнопка для запуска процесса обучения выбранных моделей. После завершения обучения должна быть сохранена наилучшая модель, и путь к ней, а также её название должны быть возвращены пользователю.  
     - **test_button**: Кнопка для тестирования новой модели на введенных текстах. После нажатия кнопки результаты классификации должны отображаться в поле `results_new_texts`.

9. **best_model_name**:
   - **Тип**: `CharField`
   - **Описание**: Поле для отображения названия наилучшей модели после завершения обучения.
   - **Виджет**: `TextInput` (только для чтения).
   - **Начальное значение**: Пустое поле.

10. **new_texts**:
    - **Тип**: `CharField`
    - **Виджет**: `Textarea`
    - **Описание**: Поле для ввода текстов, которые необходимо классифицировать.
    - **Начальное значение**: `"NASA announced a new mission to Mars\nQuantum physics breakthrough"`.

11. **results_new_texts**:
    - **Тип**: `CharField`
    - **Виджет**: `Textarea`
    - **Описание**: Поле для отображения результатов классификации новых текстов.

12. **result_html**:
    - **Тип**: `CharField`
    - **Виджет**: `Textarea`
    - **Описание**: Поле для отображения результатов классификации в формате HTML (например, для красивого оформления таблиц или графиков).



### Логика работы формы:

1. Пользователь указывает путь к датасету или выбирает название датасета из списка.
2. Пользователь выбирает категории данных, размер выборки, модели, метод векторизации и указывает, нужно ли выполнять подбор гиперпараметров.
3. После нажатия на кнопку **train_button**:
   - Запускается процесс обучения выбранных моделей.
   - Сохраняется наилучшая модель, и путь к ней, а также её название возвращаются пользователю. Название модели отображается в поле **best_model_name**.
4. Пользователь вводит тексты в поле **new_texts** и нажимает на кнопку **test_button**:
   - Тексты классифицируются с использованием наилучшей модели.
   - Результаты классификации отображаются в поле **results_new_texts**.
   - Дополнительно результаты могут быть отображены в HTML-формате в поле **result_html**.

### Форма 3: TextLemmatizationForm

**Описание**:  
Форма предназначена для выполнения лемматизации и других методов обработки текста. Пользователь может ввести текст, выбрать язык и методы обработки, а затем получить результат в текстовом и HTML-формате.

#### Поля формы:

1. **text**:
   - **Тип**: `CharField`
   - **Виджет**: `Textarea`
   - **Описание**: Поле для ввода текста, который нужно обработать.
   - **Начальное значение**: Пустое поле.

2. **language**:
   - **Тип**: `ChoiceField`
   - **Описание**: Поле для выбора языка текста. Доступные варианты: `Russian` и `English`.
   - **Начальное значение**: `Russian`.

3. **methods**:
   - **Тип**: `MultipleChoiceField`
   - **Виджет**: `CheckboxSelectMultiple`
   - **Описание**: Поле для выбора методов обработки текста из предопределенного списка (`METHOD_CHOICES`). Например: `lemmatization`, `stemming`, `stopwords_removal`.
   - **Начальное значение**: `['lemmatization']`.

4. **result_html**:
   - **Тип**: `CharField`
   - **Виджет**: `Textarea`
   - **Описание**: Поле для отображения результатов обработки текста в HTML-формате (например, для красивого оформления таблиц или выделения изменений в тексте).



### Логика работы формы:

1. Пользователь вводит текст в поле **text**.
2. Пользователь выбирает язык текста в поле **language**.
3. Пользователь выбирает методы обработки текста в поле **methods**.
4. После нажатия на кнопку обработки:
   - Текст обрабатывается выбранными методами.
   - Результаты обработки отображаются в поле **result_html** в HTML-формате.



### Форма 4: TextVectorizationForm

**Описание**:  
Форма предназначена для векторизации текста и его визуализации. Пользователь может ввести текст, выбрать язык, методы векторизации и параметры визуализации, а затем получить результат в виде графика и в HTML-формате.

#### Поля формы:

1. **text**:
   - **Тип**: `CharField`
   - **Виджет**: `Textarea`
   - **Описание**: Поле для ввода текста, который нужно векторизовать.
   - **Начальное значение**: `'Пример текста на русском. Обработка естественного языка. Машинное обучение.'`.

2. **language**:
   - **Тип**: `ChoiceField`
   - **Описание**: Поле для выбора языка текста. Доступные варианты: `Russian` и `English`.
   - **Начальное значение**: `Russian`.

3. **vismethod**:
   - **Тип**: `ChoiceField`
   - **Описание**: Поле для выбора метода визуализации. Доступные варианты: `PCA` и `UMAP`.
   - **Начальное значение**: `PCA`.

4. **n_components**:
   - **Тип**: `ChoiceField`
   - **Описание**: Поле для выбора типа графика. Доступные варианты: `2D` и `3D`.
   - **Начальное значение**: `2D`.

5. **methods**:
   - **Тип**: `MultipleChoiceField`
   - **Виджет**: `CheckboxSelectMultiple`
   - **Описание**: Поле для выбора методов векторизации из предопределенного списка (`VECTORIZATION_METHOD_CHOICES`). Например: `tf_idf`, `word2vec`, `fasttext`.
   - **Начальное значение**: `['tf_idf']`.

6. **result_html**:
   - **Тип**: `CharField`
   - **Виджет**: `Textarea`
   - **Описание**: Поле для отображения результатов визуализации в HTML-формате (например, для встраивания графиков или интерактивных визуализаций).



### Логика работы формы:

1. Пользователь вводит текст в поле **text**.
2. Пользователь выбирает язык текста в поле **language**.
3. Пользователь выбирает метод визуализации в поле **vismethod** и тип графика в поле **n_components**.
4. Пользователь выбирает методы векторизации в поле **methods**.
5. После нажатия на кнопку обработки:
   - Текст векторизуется выбранными методами.
   - Результаты векторизации визуализируются с использованием выбранного метода и типа графика.
   - Результаты визуализации отображаются в поле **result_html** в HTML-формате.




### Форма 4: UniversalCorpusForm

**Описание**:  
Форма предназначена для создания корпуса текстов из различных источников: папки с файлами, архива или веб-страниц. Пользователь может выбрать тип обработки, указать источник данных, выбрать формат выходного файла и получить путь к созданному корпусу. После заполнения всех полей и выбора параметров пользователь нажимает кнопку для создания корпуса.

#### Поля формы:

1. **process_type**:
   - **Тип**: `ChoiceField`
   - **Описание**: Поле для выбора типа обработки. Доступные варианты:
     - `Обработать из папки`
     - `Обработать из архива`
     - `Обработать веб-страницы`
   - **Начальное значение**: `Обработать из папки`.

2. **folder_path**:
   - **Тип**: `CharField`
   - **Описание**: Поле для указания пути к папке с файлами, из которых будет создаваться корпус.
   - **По умолчанию**: Скрыто (отображается только при выборе `Обработать из папки`).

3. **web_urls**:
   - **Тип**: `CharField`
   - **Виджет**: `Textarea`
   - **Описание**: Поле для ввода URL-адресов веб-страниц, которые нужно обработать для создания корпуса.
   - **По умолчанию**: Скрыто (отображается только при выборе `Обработать веб-страницы`).

4. **archive_name**:
   - **Тип**: `CharField`
   - **Виджет**: `TextInput`
   - **Описание**: Поле для указания пути к архиву, из которого будет создаваться корпус.
   - **По умолчанию**: Скрыто (отображается только при выборе `Обработать из архива`).

5. **Type_outputcorpus**:
   - **Тип**: `ChoiceField`
   - **Описание**: Поле для выбора формата выходного файла корпуса. Доступные варианты:
     - `Json`
     - `txt`
     - `xml`
     - `rtf`
     - `zip archive`
   - **Начальное значение**: `txt`.

6. **Outputcorpus_path**:
   - **Тип**: `CharField`
   - **Описание**: Поле для отображения пути к созданному корпусу.
   - **Начальное значение**: `корневой каталог программы` (например, `./output_corpus.txt`).
   - **Виджет**: `TextInput` (только для чтения).

7. **create_corpus_button**:
   - **Тип**: `Button`
   - **Описание**: Кнопка для запуска процесса создания корпуса. Доступна только после заполнения всех обязательных полей.
   - **Текст кнопки**: `Создать корпус`.



### Логика работы формы:

1. Пользователь выбирает тип обработки в поле **process_type**:
   - Если выбран `Обработать из папки`, отображается поле **folder_path**.
   - Если выбран `Обработать из архива`, отображается поле **archive_name**.
   - Если выбран `Обработать веб-страницы`, отображается поле **web_urls**.

2. Пользователь указывает источник данных (папку, архив или URL-адреса) в соответствующем поле.

3. Пользователь выбирает формат выходного файла корпуса в поле **Type_outputcorpus**.

4. После заполнения всех обязательных полей кнопка **create_corpus_button** становится активной.

5. Пользователь нажимает кнопку **create_corpus_button**:
   - Данные из указанного источника обрабатываются и создается корпус.
   - Путь к созданному корпусу отображается в поле **Outputcorpus_path**.



### Валидация:

1. Если выбран `Обработать из папки` в поле **process_type**, то:
   - Поле **folder_path** становится обязательным для заполнения.
   - Поля **web_urls** и **archive_name** скрываются.

2. Если выбран `Обработать из архива` в поле **process_type**, то:
   - Поле **archive_name** становится обязательным для заполнения.
   - Поля **folder_path** и **web_urls** скрываются.

3. Если выбран `Обработать веб-страницы` в поле **process_type**, то:
   - Поле **web_urls** становится обязательным для заполнения.
   - Поля **folder_path** и **archive_name** скрываются.

4. Кнопка **create_corpus_button** активируется только после заполнения всех обязательных полей.






### Форма 5: SocialMediaDataCollectionForm

#### **Описание:**

### **Техническое задание для реализации JWT-регистрации и аутентификации с использованием SQLite в Django**



#### **Описание:**
JWT (JSON Web Token) — это стандартный метод аутентификации, который позволяет безопасно передавать информацию между клиентом и сервером. В данном случае мы реализуем регистрацию и аутентификацию пользователей с использованием JWT, где данные будут храниться в базе данных SQLite. Реализация включает создание API-эндпоинтов для регистрации, входа, обновления токенов и выхода, а также обеспечение безопасности и удобства использования.



### **1. Архитектура системы**

1. **База данных:**  
   - Используется SQLite для хранения данных пользователей.  
   - Таблица `User` будет содержать следующие поля:  
     - `id` (первичный ключ, автоинкремент).  
     - `username` (уникальное имя пользователя).  
     - `email` (уникальный адрес электронной почты).  
     - `password` (хэшированный пароль с использованием библиотеки `bcrypt` или аналогичной).  
     - `is_active` (флаг активности пользователя, по умолчанию `True`).  
     - `date_joined` (дата и время регистрации пользователя).  

2. **JWT-токены:**  
   - При успешной аутентификации пользователь получает JWT-токен.  
   - Токен состоит из трех частей: заголовка, полезной нагрузки (payload) и подписи.  
   - Полезная нагрузка содержит информацию о пользователе (например, `user_id`, `username`, `email`).  
   - Используются два типа токенов:  
     - `access_token` — для доступа к защищенным ресурсам (срок действия: 30 минут).  
     - `refresh_token` — для обновления `access_token` (срок действия: 1 день).  

3. **API-эндпоинты:**  
   - `/api/register/` — регистрация нового пользователя.  
   - `/api/login/` — вход пользователя и получение JWT-токенов.  
   - `/api/refresh/` — обновление токена.  
   - `/api/logout/` — выход пользователя (инвалидация токена).  



### **2. Реализация форм и API**

#### **2.1. Регистрация (Registration)**

**Описание:**  
Форма регистрации создает нового пользователя и сохраняет его данные в базе данных SQLite. После успешной регистрации пользователь автоматически авторизуется.

**Поля формы:**

1. **username**  
   - **Тип**: `CharField`  
   - **Описание**: Уникальное имя пользователя.  
   - **Валидация**:  
     - Обязательное поле.  
     - Должно быть уникальным.  
     - Минимальная длина: 3 символа.  
     - Максимальная длина: 30 символов.  
     - Разрешены только буквы, цифры и символы `_`, `.`, `-`.  

2. **email**  
   - **Тип**: `EmailField`  
   - **Описание**: Уникальный адрес электронной почты.  
   - **Валидация**:  
     - Обязательное поле.  
     - Должен быть уникальным.  
     - Должен соответствовать формату email.  

3. **password1**  
   - **Тип**: `CharField`  
   - **Виджет**: `PasswordInput`  
   - **Описание**: Пароль пользователя.  
   - **Валидация**:  
     - Обязательное поле.  
     - Минимальная длина: 8 символов.  
     - Должен содержать хотя бы одну букву и одну цифру.  

4. **password2**  
   - **Тип**: `CharField`  
   - **Виджет**: `PasswordInput`  
   - **Описание**: Подтверждение пароля.  
   - **Валидация**:  
     - Обязательное поле.  
     - Значение должно совпадать с `password1`.  

**Дополнительные замечания:**  
- После успешной регистрации пользователь автоматически авторизуется и получает `access_token` и `refresh_token`.  
- На указанный адрес электронной почты отправляется письмо с подтверждением регистрации (опционально).  



#### **2.2. Вход (Login)**

**Описание:**  
Форма входа проверяет учетные данные пользователя и выдает JWT-токен при успешной аутентификации.

**Поля формы:**

1. **username_or_email**  
   - **Тип**: `CharField`  
   - **Описание**: Имя пользователя или электронная почта.  
   - **Валидация**:  
     - Обязательное поле.  
     - Должно соответствовать существующему пользователю.  

2. **password**  
   - **Тип**: `CharField`  
   - **Виджет**: `PasswordInput`  
   - **Описание**: Пароль пользователя.  
   - **Валидация**:  
     - Обязательное поле.  
     - Проверяется на соответствие учетной записи.  

**Результат:**  
- При успешной аутентификации пользователь получает два токена:  
  - `access_token` — для доступа к защищенным ресурсам.  
  - `refresh_token` — для обновления `access_token`.  



#### **2.3. Обновление токена (Refresh Token)**

**Описание:**  
Пользователь может обновить свой `access_token`, используя `refresh_token`.

**Поля запроса:**

1. **refresh_token**  
   - **Тип**: `CharField`  
   - **Описание**: Токен обновления, выданный при входе.  
   - **Валидация**:  
     - Обязательное поле.  
     - Должен быть действительным.  

**Результат:**  
- Новый `access_token` выдается пользователю.  



#### **2.4. Выход (Logout)**

**Описание:**  
Пользователь может выйти из системы, что приведет к инвалидации текущего `refresh_token`.

**Поля запроса:**

1. **refresh_token**  
   - **Тип**: `CharField`  
   - **Описание**: Токен обновления, который нужно аннулировать.  
   - **Валидация**:  
     - Обязательное поле.  

**Результат:**  
- Текущий `refresh_token` добавляется в черный список.  



### **3. База данных SQLite**

**Модель пользователя:**  
- Используется модель `User`, которая включает следующие поля:  
  - `id` (первичный ключ, автоинкремент).  
  - `username` (уникальное имя пользователя).  
  - `email` (уникальный адрес электронной почты).  
  - `password` (хэшированный пароль с использованием библиотеки `bcrypt` или аналогичной).  
  - `is_active` (флаг активности пользователя, по умолчанию `True`).  
  - `date_joined` (дата и время регистрации пользователя).  

**Хранение данных:**  
- Все данные пользователей хранятся в таблице SQLite.  
- Пароли хранятся в зашифрованном виде.  



### **4. Безопасность**

1. **Защита от CSRF:**  
   - Защита от атак Cross-Site Request Forgery обеспечивается через токены CSRF.  

2. **Ограничение количества попыток входа:**  
   - Ограничение количества неудачных попыток входа (например, 5 попыток за 5 минут).  

3. **Инвалидация токенов:**  
   - При выходе пользователя `refresh_token` добавляется в черный список.  

4. **Срок действия токенов:**  
   - `access_token` действителен 30 минут.  
   - `refresh_token` действителен 1 день.  



### **5. API-эндпоинты**

1. **POST /api/register/**  
   - Создает нового пользователя.  
   - Возвращает сообщение об успешной регистрации и токены.  

2. **POST /api/login/**  
   - Аутентифицирует пользователя.  
   - Возвращает `access_token` и `refresh_token`.  

3. **POST /api/refresh/**  
   - Обновляет `access_token` с использованием `refresh_token`.  

4. **POST /api/logout/**  
   - Инвалидирует текущий `refresh_token`.
